## Готовим датасет со ссылками для добавления новых данных

```python
import numpy as np
import pandas as pd
import time
```

```python
names = ['id',
         'header',
         'date',
         'newspaper',
         'city',
         'md_index',
         'url']

data = pd.read_excel('urls-for-parsing.xlsx',
                     names = names)
data
```

```python
cols = ['headline', 
        'description', 
        'article_body', 
        'url_parsed']

for col in cols[::-1]:
    arr = np.empty(data.shape[0])
    arr[:] = np.nan
    data.insert(2, col, arr)
    
data
```

## Импортируем тексты для добавления в датасет

```python
texts = pd.read_excel('texts-parsed.xlsx',
                      index_col=0,
                      names=cols)
texts_api = pd.read_excel('texts-parsed-api.xlsx',
                          index_col=0,
                          names=cols)
texts_all = pd.concat([texts, texts_api])
texts_all = texts_all.drop_duplicates('headline')
texts_all = texts_all.set_index(np.arange(texts_all.shape[0]))
texts_all
```

## Собираем датасет

```python
data = data.drop_duplicates(subset=['url'])
data
```

```python
links = set(texts_all.url_parsed)
rows = []

for index, row in data.iterrows():
    if row.url in list(texts_all.url_parsed):
        rows.append(row)
```

```python
data = pd.DataFrame(rows, columns = data.columns)
```

```python
# Упорядочиваем по общей переменной
data = data.sort_values(by = 'url').reset_index(drop=True)
texts_all = texts_all.sort_values(by = 'url_parsed').reset_index(drop=True)

# Склеиваем
data[cols] = texts_all[cols]
data.to_excel('temp.xlsx')

# Сортировка по дате
data = data.sort_values(by = 'date')

#Удаляем лишнее
data = data.drop(columns = ['url', 'header'])
data = data[~data.article_body.isna()]
data = data.reset_index(drop = True)

# Проставляем айди публикации
data.id = np.arange(1, data.shape[0]+1)
```

```python
urls = data.url_parsed
data = data.drop(columns=['url_parsed'])
data = pd.concat([data, urls], axis=1)
```
